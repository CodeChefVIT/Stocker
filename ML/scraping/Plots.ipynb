```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
plt.style.use('fivethirtyeight')
data = pd.read_csv("daily_top_gainers.csv")

data
```

```python
data.describe()
```

```python
ax = data.plot.bar(x = 'Company', y = '% Change', title = 'Daily Top Gainers', color='Black')
plt.savefig('daily_top_gainers.png', bbox_inches='tight')
```
